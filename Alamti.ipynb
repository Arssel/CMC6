## Матрица расстояний

```python
coords_list = []
for date in df_cut['Date'].unique():
    coords_list.append(np.array(list(map(lambda x: (float(x.split(', ')[1][:-1]),
                    float(x.split(', ')[0][1:])), df_cut[df_cut['Date'] == date]['coords'].to_list()))))
```

```python
distances_list = []
```

```python
i = 0
for coords in coords_list:
    if i < 173:
        i += 1
        continue
    n = len(coords)
    list_id = ['name' + str(i) for i in range(n)]
    if n > 1:
        r = osrm.table(coords,ids_origin=list_id, output='dataframe')
    else:
        r = None
    distances_list.append(r)
```

```python
with open('distances_list.pkl', 'wb') as f:
    pickle.dump(distances_list, f)
```
