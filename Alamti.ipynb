```python
depot = np.array([[76.988360, 43.295950]])
```

```python
dataset_size = 1
problem_size = 10000
list_id = ['name' + str(i) for i in range(problem_size + 2)]
```

```python
%%time

dataset = []
for i in range(dataset_size):
    data = {}
    rand_ind = np.random.choice(df_cut.shape[0], size=(problem_size), replace=False)
    data['demands'] = np.concatenate((np.array([0]), df_cut.iloc[rand_ind, 0].to_numpy().squeeze(), np.array([0])))
    data['coords'] = np.array(list(map(lambda x: (float(x.split(', ')[1][:-1]),
                    float(x.split(', ')[0][1:])), df_cut.iloc[rand_ind, 7].to_list())))
    data['coords'] = np.concatenate((depot, data['coords'], depot))
    data['num_vehicles'] = 12
    data['vehicle_capacities'] = [700]*data['num_vehicles']
    t = np.random.choice([0, 120, 240, 360, 480, 600], size=(problem_size + 2))
    t1 = t + 120
    data['time_windows'] = np.concatenate((t.reshape(-1, 1), t1.reshape(-1, 1)), axis=1)
    data['time_windows'][[0, problem_size+1], 0] = 0
    data['time_windows'][0, 1] = 150
    data['time_windows'][ problem_size+1, 1] = 720
    r = osrm.table(data['coords'], ids_origin=list_id, output='dataframe')
    lon = (data['coords'][:, 0] - data['coords'][:, 0].min()).reshape(-1, 1)
    lat = (data['coords'][:, 1] - data['coords'][:, 1].min()).reshape(-1, 1)
    data['coords'] = np.concatenate((lon/lon.max(), lat/lat.max()), axis=1)
    data['time_matrix'] = (r[0]/60).to_numpy()
    data['time_matrix'] += 15
    data['time_matrix'] -= np.diag([15]*(problem_size+2))
    data['time_matrix'][:, -1] = 0
    data['time_matrix'][-1, :] = 0
    dataset.append(data)
```

```python
with open('cvrp_tw_dataset.pkl', 'wb') as f:
    pickle.dump(dataset, f)
```
